```python
import os
from PIL import Image
import pandas as pd 
import glob
import random as r
import tqdm
import json
```

```python
jimmy_labels = pd.read_csv("jimmy_labels.csv")
fct_train = pd.read_csv("fcts_bboxes_processed.csv")
fct_val = pd.read_csv("lynker_val.csv")
```

```python
display(jimmy_labels.head())
display(fct_train.head())
display(fct_val.head())
```

```python
fct_train["filename"] = fct_train.filename.apply(add_fct_path)
fct_val["filename"] = fct_val.filename.apply(add_fct_path)

display(fct_train.head())
display(fct_val.head())
```

```python
def add_fct_path(fn):
    path = "rockfish/OCNMS_port"
    return os.path.join(path, fn)
```

```python
fct_train = fct_train.drop(["bbox_id", "is_validation"], axis=1)
fct_val = fct_val.drop(["bbox_id", "is_validation"], axis=1)
```

```python
merged = pd.concat([fct_train, fct_val, jimmy_labels])
merged
```

```python
merged = merged.reset_index(drop=True)
merged["bbox_id"] = merged.reset_index().index
merged
```

```python
uniques = pd.Series(merged.filename.unique())
val_images = uniques.sample(int(len(uniques)*0.2), random_state=42)
# val_index = merged.filename.sample(int(len(merged)*0.2), random_state=42)
# val_index
val_images
```

```python
merged['is_validation'] = False
```

```python
val_index = merged[merged.filename.isin(val_images)].index

merged.loc[val_index, "is_validation"] = True
merged
```

```python
merged.is_validation.value_counts()
```

```python
merged.to_csv("jimmy_fct_lynker_merged.csv", index=False)
```
