```python
boxes_list = []
lynker_guys = [david, harsh, michael]

for guy in lynker_guys:
    for key in guy:
        fn = guy[key]["filename"]
        filename = "rockfish/OCNMS_port/" + os.path.splitext(fn)[0] + ".tif" # I want .tifs instead 
        bboxes = guy[key]['regions']
        for box in bboxes:
            box = box['shape_attributes']
            x_min = box['x']
            y_min = box['y']
            x_max = box['x'] + box['width']
            y_max = box['y'] + box['height']
            bbox_dict = {
                "filename": filename,
                "x_min": x_min,
                "x_max": x_max,
                "y_min": y_min,
                "y_max": y_max,
                "is_validation": False
            }
            boxes_list.append(bbox_dict)
        
lynker_boxes = pd.DataFrame(boxes_list)
```

## take jimmy's new label jsons and reconvert to csv

```python
def coco_to_csv(coco):
    annos = coco['annotations']
    coco_dicts = []
    for anno in annos:
        x_min = anno['bbox'][0]
        y_min = anno['bbox'][1]
        x_max = anno['bbox'][2]
        y_max = anno['bbox'][3]   
        filename = anno["image_id"]
        coco_dict = {
            "x_min": x_min,
            "y_min": y_min,
            "x_max": x_max,
            "y_max": y_max,
            "filename": filename
        }
        coco_dicts.append(coco_dict)
        
    return pd.DataFrame(coco_dicts)
```

```python
jimmy = pd.concat([coco_to_csv(jimmy_train), coco_to_csv(jimmy_val)])

jimmy
```

```python
def remove_fct_path(fn):
    dirname = os.path.dirname(fn)
    return dirname
    
```

```python
from PIL import Image

def height(img):
    image = Image.open(img)
    return image.size[1]

def width(img):
    image = Image.open(img)
    return image.size[0]
```

```python
train_subset = no_fct_merge[no_fct_merge.is_validation == False].filename.unique()
```

```python
import os
from tqdm import tqdm
imgs_path = "../../data"
train_images = []
for i, img in tqdm(enumerate(train_subset)):
    train_images.append({ 
        "file_name": img,
         "height": height(os.path.join(imgs_path, img)),
         "width": width(os.path.join(imgs_path, img)),
         "id": img,
       })
print(train_images[0])
train_coco['images'] = train_images
```

```python
val_subset = no_fct_merge[no_fct_merge.is_validation == True].filename.unique()
val_images = []
for i, img in tqdm(enumerate(val_subset)):
    val_images.append({ 
        "file_name": img,
         "height": height(os.path.join(imgs_path, img)),
         "width": width(os.path.join(imgs_path, img)),
         "id": img,
       })
print(val_images[0])
val_coco['images'] = val_images
```

```python
# coco bounding box format is [top left x position, top left y position, width, height]
# TODO clip
def get_coco_bbox(xmin, xmax, ymin, ymax):
    return [ xmin, ymin, (xmax - xmin), (ymax - ymin) ]

# TODO clip, prevent negative
# divide by 2 like in camera traps example
def get_coco_area(xmin, xmax, ymin, ymax):
    return (xmax-xmin)*(ymax-ymin) / 2
```

```python
train_annotations = []
val_annotations = []
x_min = 'x_min'
y_min = 'y_min'
x_max = 'x_max'
y_max = 'y_max'
img_id = 'filename'

from tqdm import tqdm 
for index, row in tqdm(no_fct_merge.iterrows()):
    annotation = {
        "bbox": get_coco_bbox(row[x_min], row[x_max], row[y_min], row[y_max]),
        "area": get_coco_area(row[x_min], row[x_max], row[y_min], row[y_max]),
        "iscrowd": 0, # TODO: some of these ARE crowds!
        "image_id": row[img_id],
        "category_id": 1, # always a FISH
        "id": int(index) # annotation ID must be an int
    }
    if (row.is_validation == False):
        train_annotations.append(annotation)
    elif (row.is_validation == True):
        val_annotations.append(annotation)

train_coco['annotations'] = train_annotations
val_coco['annotations'] = val_annotations
```
